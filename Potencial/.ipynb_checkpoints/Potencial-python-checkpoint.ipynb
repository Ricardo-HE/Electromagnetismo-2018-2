# Potencial eléctrico

El potencial eléctrico es el trabajo que debe realizar una fuerza externa para traer una carga positiva unitaria q desde el punto de referencia hasta el punto considerado, en contra de la fuerza eléctrica y a velocidad constante. Aritméticamente se expresa como el cociente: $$V = \frac{W}{q}$$

En esta libreta se calcula el potencial eléctrico de un dipolo, es decir, el potencial debido a dos cargas puntuales, el cual esta dado por la ecuación: $$V = \frac{1}{4\pi\epsilon}(\frac{q_1}{r_1}+\frac{q_2}{r_2})$$

### Funciones que se usaran

* Se carga numpy namas porque ya es ley cargarlo cuando se usa python y porque facilita la vida en el
* matplotlib.pyplot para graficar
* Circles para poner los puntos de donde estan las cargas

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
```

```python
# Carga q
q= 6e-6
# Constante epsilon
eps0 = 8.854e-12
# distancia del dipolo
d = 1.e-1
# Calculo de k y meterle otras cosas de una vez
k = 1/(4*np.pi*eps0) * q * d

# Plano cartesiano para graficar en el
X = np.linspace(-1e1, 1e1, 1000)
Y = X.copy()
X, Y = np.meshgrid(X, Y)

# Potencial eléctrico del dipolo (V)
Phi = k * X / np.hypot(X, Y)**3

fig = plt.figure()
ax = fig.add_subplot(111)
# Dibujar contornoo en los valores de phi en niveles
levels = np.array([10**pw for pw in np.linspace(0,5,20)])
levels = sorted(list(-levels) + list(levels))

ax.contour(X, Y, Phi, levels=levels, linewidths=1)

charge_colors = {True: '#aa0000', False: '#0000aa'}
# Carga positiva
ax.add_artist(Circle((d, 0), 0.05, color=charge_colors[True]))
# Carga negativa
ax.add_artist(Circle((-d, 0), 0.05, color=charge_colors[False]))
    
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
plt.show()
```
