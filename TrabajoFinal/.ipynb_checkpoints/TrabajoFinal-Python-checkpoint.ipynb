## Cargas puntuales
### Tarea 1

Una carga puntual es una carga eléctrica hipotética, de magnitud finita, contenida en un punto geométrico carente de toda dimensión, en otras palabras una carga puntual consiste en dos cuerpos con carga que son muy pequeños en comparación con la distancia que los separa. Esta suposición resulta muy práctica al resolver problemas de electrostática, pues los efectos derivados de una distribución de cargas en un espacio finito se anulan y el problema se simplifica enormemente. 

#### Jalar datos del archivo de texto

También inicializare las variables con los valores que jale del archivo

```python
file = open("Datos.txt", "r")

noCargas = int(file.readline())

cargas = np.zeros(noCargas)
coordenadas = np.zeros((noCargas, 2))
k = 9.0e9

for i in range(noCargas):
    cargas[i] = float(file.readline())
    coordenadas[i,0] = float(file.readline())
    coordenadas[i,1] = float(file.readline())
file.close()
```

#### Calcular las fuerzas de las cargas

```python
fuerzaCargas = np.zeros((noCargas,noCargas))

# Calcular las fuerzas de cada una carga con todas las demas
# Se guardaran en la diagonal superior derecha de una matriz
for i in range(noCargas):
    for j in range(i+1,noCargas):
        radio = np.sqrt((coordenadas[j,0] - coordenadas[i,0])**2 + (coordenadas[j,1] - coordenadas[i,1])**2)
        fuerzaCargas[i,j] = np.abs(k*cargas[i]*cargas[j]/radio**2)
        
print(fuerzaCargas)
```

#### Calcular direcciones y su fuerza a ella

```python
sumaFuerzasX = np.zeros(noCargas)
sumaFuerzasY = np.zeros(noCargas)
fuerzasTotales = np.zeros(noCargas)
angulosFuerza = np.zeros(noCargas)
for i in range(noCargas):
    # Primero se suman las columnas de la carga en la matriz de fuerzas
    for j in range(noCargas):
        # calcula el angulo de la carga i respecto a la carga j
        angulo = np.arctan2(coordenadas[j,1]-coordenadas[i,1], coordenadas[j,0]-coordenadas[i,0])

        if j < i:
            # sumar fuerzas en el eje X
            if (coordenadas[i,0] > coordenadas[j,0] and cargas[i]*cargas[j] < 0) or (coordenadas[i,0] < coordenadas[j,0] and cargas[i]*cargas[j] > 0):
                sumaFuerzasX[i] -= fuerzaCargas[j,i]*np.abs(np.cos(angulo))
            elif (coordenadas[i,0] > coordenadas[j,0] and cargas[i]*cargas[j] > 0) or (coordenadas[i,0] < coordenadas[j,0] and cargas[i]*cargas[j] < 0):
                sumaFuerzasX[i] += fuerzaCargas[j,i]*np.abs(np.cos(angulo))
            # sumar fuerzas en el eje Y
            if (coordenadas[i,1] > coordenadas[j,1] and cargas[i]*cargas[j] < 0) or (coordenadas[i,1] < coordenadas[j,1] and cargas[i]*cargas[j] > 0):
                sumaFuerzasY[i] -= fuerzaCargas[j,i]*np.abs(np.sin(angulo))
            elif (coordenadas[i,1] > coordenadas[j,1] and cargas[i]*cargas[j] > 0) or (coordenadas[i,1] < coordenadas[j,1] and cargas[i]*cargas[j] < 0):
                sumaFuerzasY[i] += fuerzaCargas[j,i]*np.abs(np.sin(angulo))
        else:
            break

    for j in range(i+1,noCargas):
        angulo = np.arctan2(coordenadas[j,1]-coordenadas[i,1], coordenadas[j,0]-coordenadas[i,0])

        # Sumar fuerzas en el eje X
        if (coordenadas[i,0] > coordenadas[j,0] and cargas[i]*cargas[j] < 0) or (coordenadas[i,0] < coordenadas[j,0] and cargas[i]*cargas[j] > 0):
            sumaFuerzasX[i] -= fuerzaCargas[i,j]*np.abs(np.cos(angulo))
        elif (coordenadas[i,0] > coordenadas[j,0] and cargas[i]*cargas[j] > 0) or (coordenadas[i,0] < coordenadas[j,0] and cargas[i]*cargas[j] < 0):
            sumaFuerzasX[i] += fuerzaCargas[i,j]*np.abs(np.cos(angulo))

        #Sumar fuerzas en el eje Y
        if (coordenadas[i,1] > coordenadas[j,1] and cargas[i]*cargas[j] < 0) or (coordenadas[i,1] < coordenadas[j,1] and cargas[i]*cargas[j] > 0):
            sumaFuerzasY[i] -= fuerzaCargas[i,j]*np.abs(np.sin(angulo))
        elif (coordenadas[i,1] > coordenadas[j,1] and cargas[i]*cargas[j] > 0) or (coordenadas[i,1] < coordenadas[j,1] and cargas[i]*cargas[j] < 0):
            sumaFuerzasY[i] += fuerzaCargas[i,j]*np.abs(np.sin(angulo))

    fuerzasTotales[i] = np.sqrt(sumaFuerzasX[i]**2 + sumaFuerzasY[i]**2)
    angulosFuerza[i] = np.rad2deg(np.arctan2(sumaFuerzasY[i], sumaFuerzasX[i]))


print("Fuerzas: ", fuerzasTotales)
#display("text/plain", fuerzasTotales)
print("Angulos de las direcciones: ", angulosFuerza)
#display("text/plain", angulosFuerza)
#println()
```

## Dipolo
### Tarea 2

Las líneas de campo eléctrico para dos cargas puntuales de igual magnitud pero de signos opuestos son conocidas como dipolo eléctrico, es un sistema de dos cargas de signo opuesto e igual magnitud cercanas entre sí.

Los dipolos aparecen en cuerpos aislantes dieléctricos. A diferencia de lo que ocurre en los materiales conductores, en los aislantes los electrones no son libres. Al aplicar un campo eléctrico a un dieléctrico aislante éste se polariza dando lugar a que los dipolos eléctricos se reorienten en la dirección del campo disminuyendo la intensidad de éste. 

#### Función para encontrar el campo electrico
Encuentra el vector campo electrico

```python
def E(q, r0, x, y):
    den = np.hypot(x-r0[0], y-r0[1])**3
    return q * (x - r0[0]) / den, q * (y - r0[1]) / den
```

Se prepara los puntos a calcular sus vectores para el campo electrico, se calcula el campo electrico para cada punto y se inicializan las variables

```python
# Malla  de puntos
nx, ny = 64, 64
x = np.linspace(-2, 2, nx)
y = np.linspace(-2, 2, ny)
X, Y = np.meshgrid(x, y)
```

A partir de $n$ cargas, los acomoda en un circulo unitario con una distancia igual y cargas alternadas (uno positivo y uno negativo)

```python
nq = 2**int(1)
cargas = []
for i in range(nq):
    q = i%2 * 2 - 1
    cargas.append((q, (np.cos(2*np.pi*i/nq), np.sin(2*np.pi*i/nq))))
```

Calcula los vectores del campo electrico para cada componente

```python
# Vector del campo electrico en componentes separados
Ex, Ey = np.zeros((ny, nx)), np.zeros((ny, nx))
for carga in cargas:
    ex, ey = E(*carga, x=X, y=Y)
    Ex += ex
    Ey += ey
```

#### Gráfica

```python
fig = plt.figure()
ax = fig.add_subplot(111)

color = 2 * np.log(np.hypot(Ex, Ey))
ax.streamplot(x, y, Ex, Ey, color=color, linewidth=1, cmap=plt.cm.inferno,
              density=2, arrowstyle='->', arrowsize=1.5)

colorCargas = {True: '#aa0000', False: '#0000aa'}
for q, pos in charges:
    print(pos)
    ax.add_artist(Circle(pos, 0.05, color=colorCargas[q>0]))

ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_xlim(-2,2)
ax.set_ylim(-2,2)
ax.set_aspect('equal')
plt.show()
```

## Potencial eléctrico
### Tarea 3

El potencial eléctrico es el trabajo que debe realizar una fuerza externa para traer una carga positiva unitaria q desde el punto de referencia hasta el punto considerado, en contra de la fuerza eléctrica y a velocidad constante. Aritméticamente se expresa como el cociente: $$V = \frac{W}{q}$$

En esta libreta se calcula el potencial eléctrico de un dipolo, es decir, el potencial debido a dos cargas puntuales, el cual esta dado por la ecuación: $$V = \frac{1}{4\pi\epsilon}(\frac{q_1}{r_1}+\frac{q_2}{r_2})$$

#### Inicializando variables y calculos

```python
# Carga q
q= 6e-6
# Constante epsilon
eps0 = 8.854e-12
# distancia del dipolo
d = 1.e-1
# Calculo de k y meterle otras cosas de una vez
k = 1/(4*np.pi*eps0) * q * d

# Plano cartesiano para graficar en el
X = np.linspace(-1e1, 1e1, 1000)
Y = X.copy()
X, Y = np.meshgrid(X, Y)

# Potencial eléctrico del dipolo (V)
Phi = k * X / np.hypot(X, Y)**3
```

#### Gráfica

```python
fig = plt.figure()
ax = fig.add_subplot(111)
# Dibujar contornoo en los valores de phi en niveles
niveles = np.array([10**pw for pw in np.linspace(0,5,20)])
niveles = sorted(list(-niveles) + list(niveles))

ax.contour(X, Y, Phi, levels=niveles, linewidths=1)

colorCargas = {True: '#aa0000', False: '#0000aa'}
# Carga positiva
ax.add_artist(Circle((d, 0), 0.05, color=colorCargas[True]))
# Carga negativa
ax.add_artist(Circle((-d, 0), 0.05, color=colorCargas[False]))
    
ax.set_xlim(-1,1)
ax.set_ylim(-1,1)
plt.show()
```

#### Campo magnetico
Un campo magnético es una descripción matemática de la influencia magnética de las corrientes eléctricas y de los materiales magnéticos.1​ El campo magnético en cualquier punto está especificado por dos valores, la dirección y la magnitud; de tal forma que es un campo vectorial. Específicamente, el campo magnético es un vector axial, como lo son los momentos mecánicos y los campos rotacionales. El campo magnético es más comúnmente definido en términos de la fuerza de Lorentz ejercida en cargas eléctricas.
